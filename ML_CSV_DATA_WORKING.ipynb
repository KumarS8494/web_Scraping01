## Social Network Ads DATA

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_ads/Social_Network_Ads.csv')
```

```python
df.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(df.drop('Purchased',axis=1),
                                                df['Purchased'],test_size=0.3,random_state=0)

X_train.shape , X_test.shape
```

### StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

#fit the scaler to the train set, it will learn the parameter
scaler.fit(X_train)

#transform train and test set
X_train_scale = scaler.transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train
```

```python
X_train_scale
#it is a numpy array
#so we need to convert it DataFrame
```

```python
X_train_scale = pd.DataFrame(X_train_scale, columns=X_train.columns)
X_test_scale = pd.DataFrame(X_test_scale, columns=X_test.columns)
```

```python
#after conversion data 
X_train_scale
```

```python
#Description of original dataset
np.round(X_train.describe(),1)
```

```python
#Description of train dataset
np.round(X_train_scale.describe(),1)
#After Scale mean is and Standard_dev is 1
```

### Effect of scaling

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(X_train['Age'], X_train['EstimatedSalary'])
ax1.set_title("Before scaling")

ax2.scatter(X_train_scale['Age'], X_train_scale['EstimatedSalary'],color='red')
ax2.set_title('After Scaling')
plt.show()
```

### How do I calculate PDF (probability density function)

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(18,5))

ax1.set_title("Before scaling")
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(X_train_scale['Age'], ax=ax2)
sns.kdeplot(X_train_scale['EstimatedSalary'], ax=ax2)
plt.show()
```

### Comparision of Distributions

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(18,5))

ax1.set_title("Age Distribution Before scaling")
sns.kdeplot(X_train['Age'], ax=ax1)


ax2.set_title('Age Distribution After Scaling')
sns.kdeplot(X_train_scale['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(18,5))

ax1.set_title("EstimatedSalary Distribution Before scaling")
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

ax2.set_title('EstimatedSalary Dsitribution After Scaling')
sns.kdeplot(X_train_scale['EstimatedSalary'], ax=ax2)
plt.show()
```

### Why Scaling is important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
lr_scale = LogisticRegression()
```

```python
lr.fit(X_train, Y_train)
lr_scale.fit(X_train_scale,Y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scale = lr_scale.predict(X_test_scale)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Actual',accuracy_score(Y_test,y_pred))
print('Scaled',accuracy_score(Y_test,y_pred_scale))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
dt_scale = DecisionTreeClassifier()
```

```python
dt.fit(X_train, Y_train)
dt_scale.fit(X_train_scale,Y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_scale = dt_scale.predict(X_test_scale)
```

```python
print('Actual',accuracy_score(Y_test,y_pred))
print('Scaled',accuracy_score(Y_test,y_pred_scale))
```

```python
X_ntrain_scale = pd.DataFrame(X_ntrain_scale, columns=X_ntrain.columns)
X_ntest_scale = pd.DataFrame(X_ntest_scale, columns=X_ntest.columns)
```
